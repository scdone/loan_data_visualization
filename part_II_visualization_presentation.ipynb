## Dataset Overview

> This dataset is comprised of loan and borrower data from Prosper, a peer to peer lending company. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

%matplotlib inline

sns.set()

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# loads cleaned master loan df into df from csv file created in data_wrangling_and_cleaning.ipynb
df = pd.read_csv(r"C:\Users\14802\Desktop\DATA_PROJECTS\udacity_projects\Loan_Data_Visualization\master_loan_data.csv", index_col=False, low_memory=False)
```

## How does a borrower's credit score affect their annual interest rate?

> Using the pearsonr calculation and a seaborn regression plot, I found there is a moderate negative correlation between the borrower's loan interest rate and their credit score. As the borrower's credit score increased, their annual interest rate decreases. 

>The correlation coefficient was -0.4 with a p-value of 0, which indicates the results are statistically significant.

```python
aprs_credit = df[['credit_score_range_lower', 'borrower_a_p_r']].copy()
aprs_credit.dropna(inplace=True)
# pearsonr(aprs_credit['credit_score_range_lower'], aprs_credit['borrower_a_p_r'])
```

```python
sns.set_theme(font_scale=1.5)
plt.figure(figsize=(15,8))
ax1 = sns.regplot(x=aprs_credit['credit_score_range_lower'], y=aprs_credit['borrower_a_p_r'], color="red")
ax1.set_xlabel("Borrower's Credit Score Range")
ax1.set_ylabel("Borrower's APR")
ax1.set_title("Relationship Between Borrower's APR and Borrower's Credit Score");
```

## How does a borrower's credit score affect the estimated loan risk for the lender?

> Using the pearsonr calculation and a seaborn regression plot, I found there is a moderate negative correlation between the borrower's loan interest rate and the estimated loan risk for the lender in case of loan charge off. As the borrower's credit score increased, the estimated risk of the loan decreases. 

>The correlation coefficient was -0.5 with a p-value of 0, which indicates the results are statistically significant.

```python
credit_loss = df[['estimated_loss','credit_score_range_lower']].copy()
credit_loss.dropna(inplace=True)
# pearsonr(credit_loss['estimated_loss'],credit_loss['credit_score_range_lower'])
```

```python
sns.set_theme(style="darkgrid", font_scale=1.5)
plt.figure(figsize=(15,8))
ax1 = sns.regplot(x=credit_loss['credit_score_range_lower'], y=credit_loss['estimated_loss'])
ax1.set_xlabel("Borrower's Credit Score Range")
ax1.set_ylabel("Estimated Loss on Loan Charge-Off")
ax1.set_title("Relationship Between Estimated Loss on Loan Charge-Off and Borrower's Credit Score");
```

## How do the borrower credit score distributions vary between current loans and defaulted loans?

>The borrower credit score distribution was higher on current loans and lower on defaulted loans. In fact, the mean borrower credit score on current loans ranged from 698 - 717, while the mean borrower credit score on defaulted loans ranged from 620 - 639. 

>The median borrower credit scores for current loans ranged from 700 - 719 and the median borrower scores for defaulted loans ranged from 640 to 659.



```python
loan_categories_credit = df[['credit_score_range_lower', 'credit_score_range_upper', 'loan_status']].copy()
loan_categories_credit.dropna(inplace=True)
loan_categories_credit = loan_categories_credit.query('loan_status == "Current" or loan_status == "Defaulted"')

loan_categories_credit.rename(columns={"credit_score_range_lower":"Credit Score Range (lower-range)", "credit_score_range_upper":"Credit Score Range (upper-range)", "loan_status":"Loan Status"}, inplace=True)
```

```python
ax1 = sns.FacetGrid(loan_categories_credit, col='Loan Status', height=4, aspect=2)
ax1.map(sns.boxplot, "Credit Score Range (lower-range)", order=["Defaulted", "Current"])
ax1.fig.subplots_adjust(top=0.8)
ax1.fig.suptitle("Distribution of Credit Scores on Current Loans (lower-range)");
```

```python
ax1 = sns.FacetGrid(loan_categories_credit, col='Loan Status', height=4, aspect=2)
ax1.map(sns.boxplot, "Credit Score Range (upper-range)", order=["Defaulted", "Current"])
ax1.fig.subplots_adjust(top=0.8)
ax1.fig.suptitle("Distribution of Credit Scores on Default Loans (upper-range)");
```
